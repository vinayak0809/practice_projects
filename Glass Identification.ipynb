```python
df_1= df.drop(['Type of glass'], axis = 1)
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
```

```python
IQR  = Q3 - Q1
```

```python
Threshold = 1.5
```

```python
outliers = ((df_1 < (Q1 - Threshold * IQR)) | (df_1 > (Q3 + Threshold * IQR))).any(axis=1)
```

```python
df_clean_1 = df_1[~outliers]
```

```python
data = pd.get_dummies(df, columns=['Type of glass'])
```

```python
data.head()
```

```python
data.rename(columns = {'Type of glass_1':'window' ,'Type of glass_2':'non-window'}, inplace = True)
```

```python
data.head()
```

```python
x = data.drop(['Id number','window','non-window'],axis = 1)
```

```python
columns_1 = ['window','non-window']
y = data[columns_1]
```

```python
y.head()
```

```python
y['window'] = y['window'].astype('int')
```

```python
y.drop(columns=['non-window'],axis = 1,inplace = True)
```

```python
y.head()
```

```python
# Due to avoid overfitting we use RandomOverSampler
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_ros,y_ros,test_size= 0.2, random_state = 0)
```

```python
# Training Different Machine Learning Models
```

```python
# K-Nearest Neighbours
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# K-Nearest Neighbors
Score =[]
for i in range(2,11):
    k = KNeighborsClassifier(n_neighbors = i)
    k.fit(x_train,y_train)
    score = k.score(x_test,y_test)
    Score.append(score)
print(k.score(x_train,y_train))
print(Score)        
```

```python
# Training Accuracy is 73%
```

```python
# DecisionTreeClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Decision Tree
Scores  = []
for i in range(1):
    tree = DecisionTreeClassifier(random_state = 0)
    tree.fit(x_train,y_train)
    score = tree.score(x_test,y_test)
    Scores.append(score)
print(Scores)
print(tree.score(x_train,y_train))
```

```python
# Training Accuracy is 100% and testing_accuracy is 78%
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Logistic Regression
Scores = []
for i in range(1):
    log = LogisticRegression()
    log.fit(x_train,y_train)
    score = log.score(x_test,y_test)
    Scores.append(score)
print(Scores)
print(log.score(x_train,y_train))
```

```python
#Training accuracy is 66% and testing accuracy is 57%
```

```python
# SVM
from sklearn.svm import SVC
```

```python
Scores = []
for i in range(1):
    svc = SVC(gamma = 'auto')
    svc.fit(x_train,y_train)
    score = svc.score(x_test,y_test)
    Scores.append(score)
```

```python
print(Scores)
print(svc.score(x_train,y_train))
```
